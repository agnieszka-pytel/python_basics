## Zadanie 15.2.
#### Przetestuj sposoby wypełniania wielokątów udostępniane przez argument hatch, Np. "/" ,"|" , "—", "x", "o" ,"0", " . ", "*"

```python
t = np.linspace (0, 2*np.pi, 100, endpoint=False)
x = 16*np.sin (t)**3
y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos (4*t)
h = ['*','/','|','-','x','o','0','.']
for i in range(0,len(h)):
    plt.subplot(3,3,i+1)
    plt.fill (x, y, fill = False, hatch=h[i])
    plt.axis("equal")
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
plt.show()
```

```python
plt.fill_between(x, y, color='red', alpha=0.2, hatch="*")
plt.axis("equal")
plt.show()
```

## Zadanie 15.4.
#### Przetestuj różne napisy formatujące. Narysuj wykres funkcji f ( x ) = sin(x) przy użyciu: 
1. **czerwonych linii typu „kropka-kreska”;** 
2. **zielonych trójkącików;**
3. **niebieskich linii przerywanych oraz gwiazdek.**

```python
x = np.linspace (0, 2*np.pi, 20, endpoint=False)
y = np.sin(x)

plt.subplot(311)
plt.plot(y, 'r-.')
plt.subplot(312)
plt.plot(y, 'g^')
plt.subplot(313)
plt.plot(y, 'b*-')
plt.subplots_adjust(hspace = 0.7)
plt.show()
```

## Zadanie 15.5.
#### Narysuj koło o środku w punkcie (0, 0) i promieniu 1. Powinno być ono wypełnione półprzezroczystą barwą czerwoną, mieć silnie zarysowany brzeg i nie wyglądać jak elipsa.

## Zadanie 15.6.
#### Narysuj wykres funkcji sin(x2) na przedziale \[0, 2 pi\]. Zamieść na osi Ox etykiety  postaci 0, pi/2 ,… , 2 pi. Umieść legendę o białym tle w lewym dolnym rogu.

```python
x = np.linspace(0, 2 * np.pi, 1000)
y = np.sin (x ** 2 )
plt.plot(x, y, color="blue", label= r" $ \sin (x^2) $ " )
l = plt.legend(loc="lower left", shadow=True, frameon=True)
l.get_frame().set_facecolor("white")
plt.xticks(np.linspace(0, 2 * np.pi, 5), [r"$0$", r"$\frac{\pi}{2}$", r"$\pi$", r"$\frac{3}{2}\pi$", r"$2\pi$"])
plt.show()
```

## Zadanie 15.7.
#### Narysuj wykres kołowy oraz słupkowy podsumowujące zmienną smoker z ramki danych tips .

```python
tips = sns.load_dataset("tips")
plt.subplot(121)
n = tips['smoker'].value_counts()['Yes']
rozklad = [n, len(tips)-n]
etykiety = ['Yes','No']
x = np.arange(len(rozklad))
s = 0.75
col = [('#8bd66d'),('#6f8dd9')]
plt.bar(x, rozklad, s, color=col)
plt.xticks(x, etykiety)
plt.subplot(122)
plt.pie(rozklad, labels = etykiety, colors=col)
plt.suptitle('Liczba osób palących i niepalących')
plt.show()
```

## Zadanie 15.8.
#### Utwórz wykres skrzynkowy dla procentowego udziału napiwku w całkowitej wartości rachunku na podstawie ramki danych tips.

```python
tips = sns.load_dataset("tips")
tips["tip_frac"] = tips["tip"] / tips["total_bill"]
plt.subplot(221)
sns.boxplot(tips["tip_frac"])
plt.subplot(222)
plt.boxplot(tips["tip_frac"])
plt.subplot(223)
tips["tip_frac"].plot(kind = "box")
plt.subplot(224)
tips.boxplot(column="tip_frac")
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
plt.show()
```

##  Zadanie 15.9.
#### Utwórz wykres skrzynkowy dla procentowego udziału napiwku w całkowitej wartości rachunku w rozbiciu na podgrupy generowane przez kombinacje sex i day.

```python
tips = sns.load_dataset("tips")
tips["tip_frac"] = tips["tip"] / tips["total_bill"] * 100
sns.boxplot(tips["tip_frac"], tips["day"], tips ["sex"])
tips.boxplot(column="tip_frac", by=["day", "sex"], rot = 50)
plt.show()
```
